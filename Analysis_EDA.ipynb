# **EDA** - Exploratory Data Analysis

## **Numerical EDA** (Sales)

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Sales'], bins=50, kde=True)
plt.show()
```

Observation focus:
- Right-skewed distribution
- High-frequency low-value transactions
- Few high-value orders

## Outlier Detection of Sales

```python
sns.boxplot(x=df['Sales'])
plt.show()
```

As we can see Outliers do exist in the Sales column. But we cant remove them as there can exist Products having value tending to 0 or >20000. This is considered as a genuine outlier

## Count of Orders by Segment

```python
# Count unique orders per segment
order_counts = (
    df.groupby('Segment')['Order ID']
      .nunique()
)

# Plot pie chart
plt.figure(figsize=(6,6))
plt.pie(
    order_counts,
    labels=order_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Order Distribution by Segment')
plt.axis('equal')  # makes pie a circle
plt.show()
```

This shows that maximun orders are placed by Customer Segment users.

## Sales by Category

```python
sales_by_category = (df.groupby('Category', as_index=False)['Sales'].sum())

# Plot
plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Sales', data=sales_by_category, errorbar=None)

plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()
```

All category has Equivalent sales, Technology being slightly higher than other too. Outcome: **The performance of all category products is satisfactory**

## Top 10 Products by Sales

```python
top_products = df.groupby('Product Name')['Sales'].sum().nlargest(10)
plt.figure(figsize=(8,5))
sns.barplot(x=top_products.values, y=top_products.index)
plt.title('Top 10 Products by sale')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')
plt.show()
```

From the above graph we can conclude that Product: **Canon imageCLASS 2200 Advanced Copier** has highest sales and is the **best performing product.**

## TIME-BASED INSIGHTS

```python
# created 2 helper columns
df['Order Year'] = df['Order Date'].dt.year
df['Order Month Name'] = df['Order Date'].dt.month_name()
```

```python
monthly_sales = (
    df.groupby('Order Month Name')['Sales']
      .sum()
      .reindex([
          'January','February','March','April','May','June',
          'July','August','September','October','November','December'
      ])
)

monthly_sales.plot(marker='o', figsize=(8,5))
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

The sales increased with a sharp spike after August. Month of **November** showed highest sales whereas month of **February** showed lowest sales.

```python
yearly_sales = (
    df.groupby('Order Year')['Sales']
      .sum()
      .sort_index()
)

yearly_sales.plot(kind='line', marker='o', figsize=(8,5))
plt.title('Year-over-Year Sales')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()
```

A Linear Increase sales After the year **2016.**

## CUSTOMER & SEGMENT ANALYSIS

```python
# Aggregate total sales by segment
sales_by_segment = (df.groupby('Segment', as_index=False)['Sales'].sum())

# Plot
plt.figure(figsize=(7,5))
sns.barplot(x='Segment', y='Sales', data=sales_by_segment, errorbar=None)

plt.title('Total Sales by Segment')
plt.xlabel('Segment')
plt.ylabel('Total Sales')
plt.show()
```

The **Consumer** segment has the major contribution in the Sales of the products.

```python
# Step 1: calculate order-level total
order_level = (
    df.groupby(['Segment', 'Order ID'])['Sales']
      .sum()
      .reset_index()
)

# Step 2: calculate AOV by segment
aov_by_segment = (
    order_level.groupby('Segment')['Sales']
    .mean()
    .reset_index()
)
```

```python
plt.figure(figsize=(7,5))
sns.barplot(
    x='Segment',
    y='Sales',
    data=aov_by_segment,
    errorbar=None
)

plt.title('Average Order Value by Segment')
plt.xlabel('Segment')
plt.ylabel('Average Order Value')
plt.show()
```

**Average order value (AOV)** is equal for all segments. Reason: Consumer Segment has major Sales beacause the count of orders and customers are also high. In reality, all segments are equally performing well.

## Geographical Analysis

```python
# Aggregate sales by region
sales_by_region = (
    df.groupby('Region', as_index=False)['Sales']
      .sum()
      .sort_values('Sales', ascending=False)
)

# Plot
plt.figure(figsize=(7,5))
sns.barplot(
    x='Region',
    y='Sales',
    data=sales_by_region,
    errorbar=None
)

plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

By the graph I conclude that the **West Region** of USA have the higest sales and **South Region** having lowest, since maximum orders are from customers from West Region

```python
# Top 10 states by total sales
top_states = (
    df.groupby('State')['Sales']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x=top_states.values,
    y=top_states.index,
    errorbar=None
)

plt.title('Top 10 States by Sales')
plt.xlabel('Total Sales')
plt.ylabel('State')
plt.show()
```

**New York** has exponentially higest Sales compared to other. This is because of high Population density which results in more customer count in New York Region.

```python
pivot = pd.pivot_table(
    df,
    values='Sales',
    index='Region',
    columns='Category',
    aggfunc='sum'
)

sns.heatmap(pivot, annot=True, fmt='.0f', cmap='Blues')
plt.title('Sales Heatmap: Region vs Category')
plt.show()
```

By refering to the above heat map **Western Region** has similar amount of sales in all 3 categories where total sales being the highest. The highest sales is seen in **Technology** category of the **Eastern Region**. The **South Region** is lagging behind in sales. New startegy to target the customers is required.

## Shipping and Operations

```python
# Calculate delivery duration in days
df['Delivery Days'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
df['Delivery Days'].describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Delivery Days'], bins=30, kde=True)
plt.title('Delivery Time Distribution')
plt.xlabel('Delivery Days')
plt.ylabel('Frequency')
plt.show()
```

The following Graph Explains that maximum orders are delivered within **4-5 days**. If delivery location is clode to vendor they are delivered with a day, whereas for long distance takes 6-7 days. Second Reason the Ship Mode also affects If Maximun orders are shipped using **Second Class** then 4-5 days is common.

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    x='Ship Mode',
    y='Delivery Days',
    data=df
)

plt.title('Delivery Time by Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Delivery Days')
plt.show()
```

Here we underastand the Ship Mode Perks.
- **Same Day:** Within a day
- **First Class:** Average time of 2-3 days
- **Sencond Class:** Average time of 3-4 days
- **Standard Class:** Average time of 5-6 days

## Relationships & Comparisons

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='Delivery Days',
    y='Sales',
    data=df,
    alpha=0.6
)

plt.title('Sales vs Delivery Time')
plt.xlabel('Delivery Days')
plt.ylabel('Sales')
plt.show()
```

Sales show **no clear correlation** with delivery time high and low sales occur across all delivery days.
This indicates delivery speed isnâ€™t a primary driver of order value; other factors matter more.

```python
corr = df[['Sales', 'Delivery Days']].corr()

plt.figure(figsize=(5,4))
sns.heatmap(
    corr,
    annot=True,
    cmap='Blues',
    fmt='.2f'
)

plt.title('Correlation Matrix')
plt.show()
```

Correlation between Sales and Delivery Days is essenntialy zero.
This confirms no linear relationship delivery time has virtually no impact on sales value.
